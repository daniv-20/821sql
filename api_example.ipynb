```python
alrt = src + "alerts/active"


def alerts_in_my_area(area_code: str, severity_lvl: str):
    params = {"message_type": "alert", "area": area_code, "severity": severity_lvl}
    r = httpx.get(alrt, params=params)
    print(r.json())


alerts_in_my_area("NC", "Severe")
```

```python
cnt = src + "alerts/active/count"


def alert_counts(area_code: str, code_type: str):
    r = httpx.get(cnt)
    payload = r.json()
    area = payload[code_type]
    return area[area_code]


alert_counts("AKC068", "zones")
```

```python
srcact = src + "alerts/active"


def alrt_act(message_type, area, urgency, severity, limit):
    params = {
        "message_type": message_type,
        "area": area,
        "urgency": urgency,
        "severity": severity,
        "limit": limit,
    }
    r = httpx.get(srcact)
    payload = r.json()
    print(payload.keys())
    load = payload["features"]
    return load


alrt_act("alert", "NC", "Expected", "Severe", "1")
```
